```
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from io import BytesIO
import cv2
import requests
import os
```

```
labels = ['ben_afflek',  'elton_john',  'jerry_seinfeld',  'madonna',  'mindy_kaling']
images = [3]
for i in images:
  img_path = f"/content/drive/My Drive/celebrity/unseen_data/{i}.jpg"


  from tensorflow.keras.preprocessing import image
  import numpy as np

  img = image.load_img(img_path, target_size = (150,150))

  img_tensor = image.img_to_array(img)

  img_tensor = np.expand_dims(img_tensor, axis = 0)

  img_tensor /= 255.
  classes = model.predict(img_tensor)
  print("classes",classes)
  result = np.squeeze(classes)
  print("result",result)
  result_indices = np.argmax(result)
  print("result_indices",result_indices)
  img1 = cv2.imread(img_path, cv2.IMREAD_COLOR)
  img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
  plt.axis('off')
  plt.title("{}, {:.2f}%".format(labels[result_indices], result[result_indices]*100))
  plt.imshow(img1)
```
