```
from google.colab import drive
drive.mount('/content/drive')



from tensorflow.keras.models import load_model
model = load_model("/content/drive/My Drive/beautiful_and_average_dataset/final_pretrianed_model_beautiful_vs_avg1.h5")
```

```
import math
from random import randint
predList = []

images = [5,6,7]
for i in images:
  img_path = f"/content/drive/My Drive/beautiful_and_average_dataset/my_test_data/00{i}.jpg"

  from tensorflow.keras.preprocessing import image
  import numpy as np

  img = image.load_img(img_path, target_size = (150,150))

  img_tensor = image.img_to_array(img)

  img_tensor = np.expand_dims(img_tensor, axis = 0)

  img_tensor /= 255.

  predList.append(model.predict(img_tensor))

  import matplotlib.pyplot  as plt
  plt.imshow(img_tensor[0])
  plt.show()
```

```
for pred in predList:
  if pred >= .50:
    print("Beautiful")
  else:
    print("Average")
predList
```
